```python
import zipfile
import pandas as pd
import os

# Define the path to the ZIP file and the extraction folder
zip_file_path = '/content/archive.zip'
extract_folder = '/content/'

# Extract the ZIP file
try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_folder)
    print("Files extracted successfully.")
except FileNotFoundError:
    print("Error: ZIP file not found.")
except zipfile.BadZipFile:
    print("Error: Bad ZIP file.")

# List files in the extraction folder to confirm extraction
print("Files in /content/:", os.listdir(extract_folder))

# Define the path to the extracted CSV file
csv_file_path = '/content/tmdb_5000_movies.csv'

# Load the dataset
try:
    movies_df = pd.read_csv(csv_file_path)
    print("Dataset loaded successfully.")
    print("Available columns:", movies_df.columns.tolist())
    print("Sample data:\n", movies_df.head())
except FileNotFoundError:
    print("Error: Dataset file not found.")
except pd.errors.EmptyDataError:
    print("Error: Dataset is empty.")
except pd.errors.ParserError:
    print("Error: Dataset parsing error.")
```

```python
def recommend_movies(movies_df, genre):
    # Filter movies based on the genre keyword
    genre = genre.lower()
    recommended = movies_df[movies_df['genres'].str.contains(genre, case=False, na=False)]

    if recommended.empty:
        return "Sorry, I couldn't find any movies matching that genre."
    else:
        return recommended[['title', 'genres']].head(10)

# Example usage:
user_input = "comedy"  # This would come from user input
recommended_movies = recommend_movies(movies_df, user_input)
print("Recommended movies based on your input:\n", recommended_movies)
```

```python
class MovieBot:
    def __init__(self, movies_df):
        self.movies_df = movies_df

    def get_recommendations(self, genre):
        return recommend_movies(self.movies_df, genre)

    def chat(self):
        print("Hi! I'm MovieBot. What genre of movies are you interested in?")
        while True:
            user_input = input("You: ").strip().lower()
            if user_input in ['quit', 'exit', 'bye']:
                print("MovieBot: Goodbye! Have a great day!")
                break
            else:
                recommendations = self.get_recommendations(user_input)
                print("MovieBot:", recommendations)

# Initialize and start the chatbot
bot = MovieBot(movies_df)
bot.chat()
```

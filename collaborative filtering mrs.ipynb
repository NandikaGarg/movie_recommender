```python
import numpy as np
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv(r'F:\Python34\5000_movies.csv')
credits=pd.read_csv(r"F:\Python34\5000_credits.csv")
```

```python
movies=movies.merge(credits,on='title')
movies.head(1)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(1)
```

```python
ratings=pd.read_csv(r"F:\Python34\ratings_small.csv")
scores=pd.concat([movies,ratings],axis=1,sort=False)
scores.head()
```

```python
user_scores=scores.pivot_table(index="userId",columns=['title'],values='rating')
user_scores.head(1)
```

```python
user_scores=user_scores.fillna(0)
user_scores.head(10)
```

```python
similarity=user_scores.corr(method='pearson')
similarity.head(100)
```

```python
def get_similar(movie_name,user_score):
    similarity_score=similarity[movie_name]*(user_score-2.5)
    similarity_score= similarity_score.sort_values(ascending=False)
    return similarity_score
```

```python
print(get_similar('Avatar',5))
```

```python
similar_scores=pd.DataFrame()
```

```python
robotics_lover=[('Avatar',1),('RoboCop',2)]
```

```python
for movieId,rating in robotics_lover:
   similar_scores=pd.concat([similar_scores,get_similar(movieId,rating)],ignore_index=True)
similar_scores.head()
similar_scores.sum().sort_values(ascending=False)
```
